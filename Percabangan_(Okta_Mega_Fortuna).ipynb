# **Percabangan/Struktur Pemilihan**

Pada bagian ini, kamu akan membuat keputusan pada bahasa pemrograman python menggunakan struktur percabangan atau juga sering disebut IF ELSE Statement. Pengambilan keputusan dibutuhkan ketika kamu akan mengeksekusi suatu kode jika sebuah kondisi tertentu dipenuhi.

Statement if .. elif ... else digunakan pada bahasa python untuk pengambilan keputusan.

# **IF Statements**

**Python IF Statements Syntax**

```python
if test expression :
  statement(s)
```

Program akan mengevalusi nilai dari test expression yang diberikan kemudian akan mengeksekusi statement(s) hanya jika test expression bernilai True.

Jika test expression bernilai False, maka statement(S) tidak dieksekusi.

Seperti telah disampaikan pada pertemuan sebelumnya bahwa bagian body dari statement if diindikasikan dengan indentasi. Statement-statement yang akan dijalankan jika test expression bernilai True ditulis pada baris selanjutnya dan menjorok kedalam.

Sebagai catatan tambahan bahwa python menginterpretasikan non-zero values sebagai True. None dan 0 diinterpretasikan sebagai False.

**Python IF Statement Flowchart**

Capture 8.JPG

**Example : Python IF Statement**

```python
# Jika angka adalah bilangan positif maka print "adalah bilangan positif"

angka = 3
if angka > 0 :
  print(angka, "adalah bilangan positif.")
print("Pernyataan ini selalu ditampilkan.")

angka = -1
if angka > 0 :
  print(angka, "adalah bilangan positif.")
print("Pernyataan ini juga selalu ditampilkan.")
```

Pada contoh di atas, angka > 0 adalah test expression. Bagian body dari if dieksekusi hanya jika hasil evaluasi test expression adalah bernilai True.

Ketika variable angka bernilai 3, test expression bernilai True sehingga statement di dalam bagian body dari if akan dieksekusi. Sedangankan ketika variable angka bernilai -1, test expression bernilai False sehingga statement didalam bagian body dari if tidak dieksekusi(diskip).

Statement print() diluar blok if (yang tidak terindentasi) akan selalu dieksekusi.

# **IF ELSE Statements**

**Pyton IF ELSE Statement Syntax**

```python
if test expression :
  Body of if
else :
  Body of else
```

Statement if ... else mengevaluasi test expression dan akan mengeksekusi bagian body of if hanya jika test expression bernilai True.

Jika test expression bernilai False, maka bagian body of else akan dieksekusi. Indentasi juga digunakan pada bagian ini.

**Python IF ELSE Statement Flowchart**

Capture 9.JPG

**Example : IF ELSE Statement**

```python
# Program disusun untuk mengecek apakah angka merupakan bilangan positif atau bilangan negatif
# Dan menampilkan tulisan yang sesuai

angka = 3

#Kamu bisa mencoba angka-angka lainnya berikut :
#angka = -5
#angka = 0

if angka >= 0 :
  print("Bilangan Positif atau Nol")
else :
  print("Bilangan Negatif")
```

Pada contoh diatas, ketika angka sama dengan 3, test expression bernilai True dan bagian body dari if dieksekusi sedangkan bagian body dari else diabaikan.

Ketika angka sama dengan -5, test expression bernilai False dan bagian body dari if diabaikan sedangkan bagian body dari else dieksekusi.

Jika angka sama dengan 0, test expression bernilai True dan bagian body dari if dieksekusi sedangkan bagian body dari else diabaikan.

```python
if test expression :
  Body of if
elif test expression :
  Body of elif
else :
  Body of else
```

#**Python Nested IF Statements**

Kita dapat membuat statement if...elif...else di dalam statement if...elif...else lainnya. Ini disebut sebagai nesting dalam computer programming .

Sejumlah pernyataan ini dapat bersarang( nesting ) di dalam satu sama lain. Indentasi adalah satu-satunya cara untuk mengetahui tingkat bersarang. Mereka bisa membingungkan, jadi mereka harus dihindari kecuali jika perlu.

**Example: Python NESTED IF Statement**

```python
# Program disusun untuk mengecek apakah angka merupakan bilangan positif, negatif atau nol 
# dan menampilkan tulisan yang sesuai

angka = float(input("Masukkan angka: "))
if angka >= 0:
  if angka == 0:
    print("Nol")
  else:
    print("Bilangan Positif")
else:
  print("Bilangan Negatif")
```

2. Diberi sebuah string, tulislah sebuah fungsi python untuk memeriksa apakah string tersebut merupakan
palindrom atau bukan. Sebuah string dikatakan palindrom jika susunan terbalik dari string tersebut sama
dengan string pada susunan awalnya. Misalnya, "radar" adalah palindrom, tetapi "radix" bukanlah
palindrom.

```python
a = input("Kata: ")
z = len(a)
kondisi = True
for i in range (0, z):
  if (kondisi):
    b = z-(i+1)
    if a[i] is a[b]:
      kondisi = True
    else:
      kondisi = False
if (kondisi):
  print('Palindrom')
else:
  print('Bukan Palindrom')
```

```python
daftar_harga = {'telur' : 20000, 'gula_pasir' : 16000, 'teh' : 4000}
daftar_belanjaan = {'telur' : 2, 'gula_pasir' : 5, 'teh' : 10}
def pembayaran(daftar_harga, daftar_belanjaan):
  jumlah = 0
  for barang in daftar_harga:
    for beli in daftar_belanjaan:
      if barang == beli:
        jumlah += daftar_harga[beli]*daftar_belanjaan[beli]
  disc = jumlah-jumlah*0.25
  return disc
jumlah_bayar = pembayaran(daftar_harga, daftar_belanjaan)
print(jumlah_bayar)
```
